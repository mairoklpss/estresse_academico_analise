```python
# remove um campo do df
df.drop(columns=['Timestamp'], inplace=True)
```

```python
# traduz as colunas
new_column_names = {
    'Your Academic Stage': 'Estágio Acadêmico',
    'Peer pressure': 'Pressão dos Colegas',
    'Academic pressure from your home': 'Pressão Acadêmica Familiar',
    'Study Environment': 'Ambiente de Estudo',
    'What coping strategy you use as a student?': 'Estratégia de Enfrentamento',
    'Do you have any bad habits like smoking, drinking on a daily basis?': 'Hábitos Diários (Fumo/Álcool)',
    'What would you rate the academic  competition in your student life': 'Competição Acadêmica',
    'Rate your academic stress index ': 'Índice de Estresse Acadêmico'
}
df.rename(columns=new_column_names, inplace=True)
df.head()
```

```python
df['Estágio Acadêmico'].value_counts()
```

```python
# traduz os registro de estagio
df['Estágio Acadêmico'] = df['Estágio Acadêmico'].replace({
    'undergraduate': 'graduação',
    'high school': 'ensino médio',
    'post-graduate': 'pós-graduação'
})

df['Estágio Acadêmico'].value_counts()
```

```python
# traduz os registro de ambiente
df['Ambiente de Estudo'] = df['Ambiente de Estudo'].replace({
    'Peaceful': 'tranquilo',
    'disrupted': 'perturbado',
    'Noisy': 'barulhento'
})

df['Ambiente de Estudo'].value_counts()
```

```python
df.head()
```

```python
df['Estratégia de Enfrentamento'].value_counts()
```

```python
# traduz os registro de estrategia
df['Estratégia de Enfrentamento'] = df['Estratégia de Enfrentamento'].replace({
    'Analyze the situation and handle it with intellect': 'Analisar a situação e lidar com inteligência',
    'Emotional breakdown (crying a lot)': 'Crise emocional (chorar muito)',
    'Social support (friends, family)': 'Apoio social (amigos, família)'
})

df['Estratégia de Enfrentamento'].value_counts()
```

```python
df['Hábitos Diários (Fumo/Álcool)'].value_counts()
```

```python
# traduz os registro de habitos
df['Hábitos Diários (Fumo/Álcool)'] = df['Hábitos Diários (Fumo/Álcool)'].replace({
    'No': 'não tem',
    'Yes': 'tem',
    'prefer not to say': 'prefere não dizer'
})

df['Hábitos Diários (Fumo/Álcool)'].value_counts(
)
```

```python
df.head()
```

```python
df['Pressão dos Colegas'].value_counts()
```

```python
df['Pressão Acadêmica Familiar'].value_counts()
```

```python
# traduz os registro de pressao familiar
df['Pressão Acadêmica Familiar'] = df['Pressão Acadêmica Familiar'].replace({
    1: 'Muito Baixa',
    2: 'Baixa',
    3: 'Moderada',
    4: 'Alta',
    5: 'Muito Alta'
})

df['Pressão Acadêmica Familiar'].value_counts()
```

```python
df['Competição Acadêmica'].value_counts()
```

```python
# traduz os registro de competição
df['Competição Acadêmica'] = df['Competição Acadêmica'].replace({
    1: 'Muito Baixa',
    2: 'Baixa',
    3: 'Moderada',
    4: 'Alta',
    5: 'Muito Alta'
})

df['Competição Acadêmica'].value_counts()
```

```python
df['Índice de Estresse Acadêmico'].value_counts()
```

```python
# traduz os registro de indice
df['Índice de Estresse Acadêmico'] = df['Índice de Estresse Acadêmico'].replace({
    1: 'Muito Baixa',
    2: 'Baixa',
    3: 'Moderada',
    4: 'Alta',
    5: 'Muito Alta'
})

df['Índice de Estresse Acadêmico'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
import plotly.express as px
```

```python
#colocando os dados de indice de estresse do maior para o menor
dados_maior_menor_indice = df.groupby('Índice de Estresse Acadêmico')['Índice de Estresse Acadêmico'].value_counts().sort_values(ascending=False)
dados_maior_menor_indice
```

```python
#colocando os dados de indice de estresse do menor para o maior
dados_menor_maior_indice = df.groupby('Índice de Estresse Acadêmico')['Índice de Estresse Acadêmico'].value_counts().sort_values(ascending=True)
dados_menor_maior_indice
```

```python
# criação de pizza com plotly
stress_contagem = dados_menor_maior_indice.reset_index()

#contagem dos indices de estresse academico
stress_contagem.columns = ['Índice', 'Quantidade']

#construçao da figura
fig = px.pie(stress_contagem,
             names = 'Índice',
             values = 'Quantidade',
             title = 'Proporção dos Índices de Estresse Acadêmico dos Estudantes')

fig.show()
```

```python
#tamanho do gráfico
plt.figure(figsize=(10, 8))

#construção do gráfico COM SNS. tendo como base a variável criada "dados_maior_menor".
# o .index serve para que o plt reconheça os nomes da tabela.
sns.barplot(x=dados_maior_menor_indice.index, y=dados_maior_menor_indice.values, order=dados_maior_menor_indice.index)

plt.title("Índice de Estresse Acadêmico")

plt.xlabel("Índice de Estresse Acadêmico")

plt.ylabel("Quantidade de Estudantes")

plt.show()
```

```python
# construção de um gráfico em barra COM PX.
fig = px.bar(dados_menor_maior_indice.index, #dados
             x='Índice de Estresse Acadêmico', #dados do eixo x
             y=dados_menor_maior_indice.values, #dados do eixo y
             title= "Índice de Estresse Acadêmico", #titulo do grafico
             # labels serve para nomear os rótulos x e y quando é passado o cursor em cima da barra.
             labels= {'Índice de Estresse Acadêmico': 'Nível de Estresse Acadêmico', 'y': 'Quantidade de alunos'},
             color='Índice de Estresse Acadêmico' #colorir barras
             )

# nomear os eixos.
fig.update_layout(xaxis_title='Nível de Estresse Acadêmico', yaxis_title='Quantidade de aluno')
fig.show()
```

```python
# agrupando ambiente de estudo do maior para o menor
df_ambiente_estudo_maior_menor = df.groupby('Ambiente de Estudo')['Ambiente de Estudo'].value_counts().sort_values(ascending=False)
df_ambiente_estudo_maior_menor
```

```python
fig = px.bar (
    df_ambiente_estudo_maior_menor.index,
    x= 'Ambiente de Estudo',
    y= df_ambiente_estudo_maior_menor.values,
    title= "Característica do Ambiente de Estudo dos Estudantes",
    labels= {'Ambiente de Estudo': 'Característica', 'y': 'Quantidade de alunos'},
    color= 'Ambiente de Estudo'
    )

fig.show()
```

```python
# Criando um novo DataFrame com apenas as colunas de ambiente e índice de estresse
df_ambiente_estresse = df[['Ambiente de Estudo', 'Índice de Estresse Acadêmico']]

# Exibindo as primeiras linhas do novo DataFrame
df_ambiente_estresse.head()
```

```python
# Agrupando os dados para CONTAR as ocorrências de cada combinação
# Criando um novo campo chamado "Quantidade": .size().reset_index(name='Quantidade')
# Em relaçao ao novo dataframe feito anteriormente
contagem_ambiente_indice = df_ambiente_estresse.groupby(['Ambiente de Estudo', 'Índice de Estresse Acadêmico']).size().reset_index(name='Quantidade')

# Criando o gráfico de barras agrupadas
# Índice de estresse para cada tipo de ambiente de estudo
fig = px.bar(contagem_ambiente_indice,
             x="Ambiente de Estudo",
             y="Quantidade",
             color="Índice de Estresse Acadêmico",
             barmode="group", #barmode faz com que as barras fiquem uma do lado do outra
             title="Índice de Estresse por Ambiente de Estudo",
             text_auto=True # Adicionado para exibir a quantidade exata nas barras
            )

fig.update_layout(xaxis_title= 'Característica do Ambiente de Estudo', yaxis_title= 'Quantidade de alunos')

fig.show()
```

```python
# Filtrar os dados apenas para o ambiente 'Barulhento'
# em relação a variavel criada anteriormente.
df_barulhento = contagem_ambiente_indice[contagem_ambiente_indice['Ambiente de Estudo'] == 'Barulhento']
df_barulhento.head()
```

```python
# Filtrar os dados apenas para o ambiente 'Tranquilo'
# em relação a variavel criada anteriormente.
df_tranquilo =  contagem_ambiente_indice[contagem_ambiente_indice['Ambiente de Estudo'] == 'Tranquilo']
df_tranquilo
```

```python
df_perturbado = contagem_ambiente_indice[contagem_ambiente_indice['Ambiente de Estudo'] == 'Perturbado']
df_perturbado
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

#grafico em pizza mostrando os indices de estresse em ambiente barulhento
fig1 = px.pie(
    df_barulhento,
    names= 'Índice de Estresse Acadêmico',
    values= 'Quantidade',
    title= 'Índice de Estresse Acadêmico em Ambiente de Estudo Barulhento',
    color= 'Índice de Estresse Acadêmico',
    labels= {'Índice de Estresse Acadêmico': 'Índice de Estresse'}
)

# update_traces(textinfo= 'percent+label') adiciona label+porcentagem nos setores
fig1.update_traces(textinfo= 'percent+label')


#grafico em pizza mostrando os indices de estresse em ambiente tranquilo
fig2 = px.pie(
    df_tranquilo,
    names= 'Índice de Estresse Acadêmico',
    values= 'Quantidade',
    color= 'Índice de Estresse Acadêmico',
    title= 'Índice de Estresse Acadêmico em Ambiente de Estudo Tranquilo',
)

fig2.update_traces(textinfo= 'percent+label')

#grafico em pizza mostrando os indices de estresse em ambiente perturbado
fig3 = px.pie(
    df_perturbado,
    names= 'Índice de Estresse Acadêmico',
    values= 'Quantidade',
    color= 'Índice de Estresse Acadêmico',
    title= 'Índice de Estresse Acadêmico em Ambiente de Estudo Perturbado',
)

fig3.update_traces(textinfo= 'percent+label')

# Criar subplots: 1 linha, 3 colunas
fig_combined = make_subplots(rows=1, cols=3, specs=[[{'type':'domain'}, {'type':'domain'}, {'type':'domain'}]],
                             subplot_titles=["Ambiente Barulhento", "Ambiente Tranquilo", "Ambiente Perturbado"])

# Adicionar as traces de cada figura aos subplots
fig_combined.add_trace(fig1.data[0], 1, 1)
fig_combined.add_trace(fig2.data[0], 1, 2)
fig_combined.add_trace(fig3.data[0], 1, 3)

# Atualizar layout geral
fig_combined.update_layout(title_text="Índice de Estresse Acadêmico por Ambiente de Estudo")

fig_combined.show()
```
